```r
library(tidyverse)
```

```r
ncbi<-read.table('Acropora_tenuis_NCBI.txt',header=TRUE)
```

```r
cols<-c('materialSampleID','principalInvestigator','yearCollected','decimalLatiude','decimalLongitude','country','genus','specificEpithet','lifeStage','monthCollected','dayCollected','permitInformation','tissueType','depthOfBottomInMeters')
```

```r
geome<-as.data.frame(matrix(NA,ncol=length(cols),nrow=nrow(ncbi)))
colnames(geome)<-cols
```

```r
colnames(ncbi)
colnames(geome)
```

```r
geome$materialSampleID<-ncbi$sample_name
```

```r
geome$principalInvestigator<-'Lisa McManus'
```

```r
geome$yearCollected<-2016
```

```r
ncbi$lat<-sapply(strsplit(as.character(ncbi$lat_lon), "_"), `[`, 1)
ncbi$lon<-sapply(strsplit(as.character(ncbi$lat_lon), "_"), `[`, 2)
```

```r
geome$decimalLatiude<-ncbi$lat
geome$decimalLongitude<-ncbi$lon
```

```r
geome$country<-'Philippines'
```

```r
geome$genus<-'Acropora'
```

```r
geome$specificEpithet<-'tenuis'
```

```r
geome$lifeStage<-'adult'
```

```r
ncbi$month<-sapply(strsplit(as.character(ncbi$collection_date), "-"), `[`, 2)
ncbi$day<-sapply(strsplit(as.character(ncbi$collection_date), "-"), `[`, 3)
```

```r
geome$monthCollected<-ncbi$month
geome$dayCollected<-ncbi$day
```

```r
geome$permitInformation<-'Gratuitous Permit 0129-17 and CITES export permit 4733'
```

```r
geome$tissueType<-ncbi$host_tissue_sampled
```

```r
geome$depthOfBottomInMeters<-ncbi$depth
```

```r
geome
```

```r
write.csv(geome,'Acropora_tenuis_geome.csv', row.names = FALSE)
```
